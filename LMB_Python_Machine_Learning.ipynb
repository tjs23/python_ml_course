The optimal values for these model parameters are learned from the data (this is the "learning" in machine learning), which is often called training the model.

## Quantitative data
__Quantitative data__ comes in two types: __Discrete__ and __Continuous__. As a general rule, counts are discrete and measurements are continuous.

__Discrete__ data is a count that cannot be made more precise and are represented as integers. For instance, the number of people in a room is discrete data, because you are counting whole, indivisible entities.

Eg. (1, 2, 5, 10, 20, 100)

__Continuous__ data on the other hand can be divided and reduced to higher levels of precision. For example, measuring the mass or temperature of something can be done on precise scales. Continuous data often has units of measure, but not always.

Eg. (-1.03, 4.25, 5.33)


## Qualitative data
__Qualitative data__ arises when you classify or categorise something, and comes in three types: __Binomial__, __Nominal__ and __Ordinal__.

__Binomial__ or binary data occurs when you place things in one of two mutually exclusive categories: True/False, Disease/No Disease, Good/Bad.


__Nominal__ data occurs when we assign individual items to named categories that do not have an implicit natural value or rank. There is no scale or mathematical meaning to numerical values that you might assign to the values.

Further, nominal data has no order amongst the categories.

Eg. Blood type, political party, words, nucleic acid and protein sequences.


__Ordinal__ data represents discrete and ordered units. It is very similar to nominal data, except the categories have an implicit or natural order, such as "Short, Medium or Tall". Ordinal data can be ranked.

Eg. A survey question that asks to rate an item on a 1 to 10 scale, with 10 being the best.

## Preprocessing data

### Data representation in Scikit-learn

In general, a learning problem considers a set of n __samples__ of data and then tries to predict properties (eg. __targets__) of unknown data. If each sample is more than a single number and, for instance, a multi-dimensional entry, it is said to have several attributes or __features__.

One way to think about data for machine learning with Scikit-learn is to initially think about tables (much like an excel spreadsheet).

A basic table is a two-dimensional grid of data, where rows represent individual elements of the dataset and the columns represent quantities (features and or targets) related to each of these elements.

For example, we can load up the [Iris flower dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set) and display it in the form of a table.

````python
import seaborn as sns

iris = sns.load_dataset('iris')
iris
````

In this example, each row refers to a single observed flower (__sample__), and the number of rows is the total number of __samples__ in the dataset. As such, each column of the data refers to a particular quantitative piece of information that describes each sample, and this is commonly called the __features__ of the dataset. In supervised learning, one may wish to predict a feature, such as species, for new samples. The species column would then be referred to as a __target__.

As a side note, in the above example __features__ are continuous data, and the __target__ is nominal data.

#### Feature and target matrices
Although tables are great for visualisation, Scikit-learn and essentially every other python machine learning library requires feature/target information in the form of a NumPy array/matrix. By convention, the __features__ matrix is stored in a variable called ___X___. The features matrix is often assumed to be a two-dimensional array with the shape `[n_samples, n_features]`. Features are typically real values, but may also be boolean (True/False) or discrete values (categorical) in some cases.

In addition to the feature matrix ___X___, supervised learning also requires a label or target matrix. This is typically a quantity that we want to predict from the data. By convention, the ___target___ matrix is stored in a variable called ___y___. The target matrix is usually one-dimensional, with the length of `[n_samples]`.

The target matrix can be continuous real values (regression) or discrete classes/labels (binary classification, multiclass classification). Some Scikit-learn models can handle multiple target values in a two-dimensional target array `[n_samples, n_targets]`, but you will need to check the documentation.

In the above example, we may wish to build a model that can predict the species of flower based on other measurements, in this case, the _species_ column would be the __target__ matrix.

### Representation of common biological data

Whilst the prior example Iris dataset is fairly simple to transform into feature and target matrices, other types of datasets commonly seen in biology are a bit more complex.

#### Biological sequences
For example, nucleic acid and protein sequences are not numerical, although they are discrete in nature.

When representing biological sequences, there are a few common approaches to take.

The following sequence is 16 nucleotides long, and each nucleotide is considered to be a __feature__.

>`AGTTGAGATCACAATG`

If one attempted to pass a feature matrix with characters into a Scikit-learn model for training, it would most likely throw an exception error. As it would not know how to handle the data type.

The most simplest representation would be to convert the nucleotides into discrete numerical values.


````python
## Vectorising nucleic acid sequences as discrete integers.

nt_to_int = {
    'A': 0,
    'T': 1,
    'C': 2,
    'G': 3
}

def encode_sequence_int(sequence):
    feature_array = np.zeros((len(sequence)), dtype=np.int)
    for i, nt in enumerate(sequence):
        feature_array[i] = nt_to_int[nt]
    return feature_array

sample_1 = 'AGTTGAGATCACAATG'

encoded_sample_1 = encode_sequence_int(sample_1)

print(sample_1)
print(encoded_sample_1)
````

As some machine learning algorithms require the features to be scaled between -1 and +1 or 0 and +1; another representation is to convert the sequence into a binary two-dimensional matrix. This approach is called __one hot encoding__.

````python
## Vectorising nucleic acid sequences via one hot encoding
import numpy as np

nt_to_bool = {
    'A': 0,
    'T': 1,
    'C': 2,
    'G': 3
}

def encode_sequence_one_hot(sequence):
    feature_array = np.zeros((len(sequence), 4), dtype=np.int)
    for i, nt in enumerate(sequence):
        feature_array[i, nt_to_bool[nt]] = 1
    return feature_array

sample_1 = 'AGTTGAGATCACAATG'

encoded_sample_1 = encode_sequence_one_hot(sample_1)

print(sample_1)
print(encoded_sample_1)
````

However, this poses a problem for many Scikit-learn models, where they are unable to handle multi-dimensional feature inputs.

A simple solution to this problem is to flatten the two-dimensional feature array into a one-dimensional array (_vector_) using the `.flatten()` method of a NumPy array.

````python
encoded_sample_1_flat = encoded_sample_1.flatten()
print(encoded_sample_1_flat)

print(encoded_sample_1.shape)
print(encoded_sample_1_flat.shape)
````

## Preparing your data - feature scaling

Many machine learning algorithms require features to be on the same scale, and typically between -1 and +1; and some algorithms require features to be between 0 and +1. _Decision trees and random forests are two of the few machine learning algorithms that are scale invariant._

Further, some optimization algorithms such as gradient descent work best if our features are centered at a mean of zero with a standard deviation of one (normally distributed).

There are two common approaches to feature scaling: __normalisation__ and __standardisation__.


### __1. Min-Max scaling__

Min-Max scaling, often called __normalisation__, is the simplest approach to rescaling features within a dataset. In this approach, the data is scaled to a fixed range, such as between 0 and 1. This method is most typically applied to image data, and can be calculated as follows:

$ x' = \frac{x-min(x)}{max(x)-min(x)}$

or

```python
x_scaled = (x - np.min(x)) / (np.max(x) - np.min(x)) ```

or

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
```



### __2. Standardisation__

Standardisation is considered to be a more reliable means of feature scaling, as it rescales the data such that it has the properties of a normal distribution with a mean of 0 and standard deviation of 1; thus, the feature values are not specifically bounded.

This is important for many learning algorithms in Scikit-learn, where weights are typically initialised with a value of 0. Further, standardisation retains useful information about outliers, and makes learning less sensative to them in comparison with min-max scaling.

Standardisation is calculated as follows:

$ x' = \frac{x-mean(x)}{std(x)}$

or

```python
x_scaled = (x - np.mean(x)) / np.std(x)```

or

```python
from sklearn import preprocessing
standard_scaler = preprocessing.StandardScaler()
X_scaled = standard_scaler.fit_transform(X)
```


These are not the only feature scaling methods used in the literature.
Scikit-learn [implements these and several more advanced methods](https://scikit-learn.org/stable/modules/preprocessing.html), explaining when they should/should not be used, along with their [benefits and weaknesses](https://scikit-learn.org/stable/auto_examples/preprocessing/plot_all_scaling.html#sphx-glr-auto-examples-preprocessing-plot-all-scaling-py).


````python
x = np.array([3,4,5,6,7,7,7,8,8,8,8,8,9,9,9,10,10,11,12,15,18,25])

fig, ax = plt.subplots(1,3, figsize=(12,4))
ax[0].hist(x, bins=25)
ax[0].set_title("Raw data")


ex_minmax = (x-np.min(x)) / (np.max(x)-np.min(x))
ax[1].hist(ex_minmax, bins=25)
ax[1].set_title("Min-Max scaling")

ex_stand = (x - np.mean(x)) / np.std(x)
ax[2].hist(ex_stand, bins=25)
ax[2].set_title("Standardisation")

plt.show()
````

## Transformations

The above examples purely scale the data between different ranges, they do not alter the relative distance between the features, nor do they alter the distribution of the data. That is, they perform a linear shift of the data.

Transformers perform non-linear transformations in which distances between marginal outliers and inliers are shrunk. However, it does distort correlations and distances within and across features.

A common and very simple approach is to log transform your dataset before scaling it, as seen below.

Scikit-learn provides a number of more [powerful transformation methods](https://scikit-learn.org/stable/auto_examples/preprocessing/plot_all_scaling.html).

````python
x = np.array([3,4,5,6,7,7,7,8,8,8,8,8,9,9,9,10,10,11,12,15,18,25])
x = np.log(x)

fig, ax = plt.subplots(1,3, figsize=(12,4))
ax[0].hist(x, bins=25)
ax[0].set_title("Log transformed data")

ex_minmax = (x-np.min(x)) / (np.max(x)-np.min(x))
ax[1].hist(ex_minmax, bins=25)
ax[1].set_title("Log Min-Max scaling")

ex_stand = (x - np.mean(x)) / np.std(x)
ax[2].hist(ex_stand, bins=25)
ax[2].set_title("Log standardisation")

plt.show()
````

## How and when to scale/transform

### The train/test split

When training a machine learning model, one often wishes to determine whether the model has overfit to the training data, or has generalised well for predicting unseen data.

To do this, we typically split 10-30% of the initial dataset via random selection and set it aside.

This results in what is called the ___train/test split___.

Scikit-learn provides a convenient function, `train_test_split` for performing this split on a given dataset.

````python
from sklearn.model_selection import train_test_split

X, y = np.arange(10), np.arange(10)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(X)

print(X_train)

print(X_test)
````

### __A common question is how do we scale our data correctly?__

* __At what point should we scale the data?__
* __How do we scale new unseen data to be in the same range as the training data?__


We have several options available to us. Can you guess the correct one?


### Scenario 1:

```python
scaled_dataset = (dataset - dataset_mean) / dataset_std_deviation
train, test = train_test_split(scaled_dataset)
```

### Scenario 2:

```python
train, test = train_test_split(dataset)
scaled_train =  (train - train_mean) / train_std_deviation
scaled_test = (test - test_mean) / test_std_deviation
```

### Scenario 3:

```python
train, test = train_test_split(dataset)
scaled_train =  (train - train_mean) / train_std_deviation
scaled_test = (test - train_mean) / train_std_deviation
```

__Scenario 3__ is the correct approach.

Whilst it may look strange to reuse the training dataset mean and standard deviation, we do so for good reason.

When encountering unseen data, we need to scale its features correctly, and often, the unseen data will be a single sample.

In that case, it is not possible to calculate a mean and standard deviation for scaling.

Further, during training, we wish to treat the testing dataset as _unseen_ data.


__To recapitulate:__ If we scale and or transform our training dataset, we need to keep and store the parameters (eg. mean and standard deviation) for each feature. Then use these parameters when preprocessing our data and making predictions in the future.

### Notes on sparse data

The previously explored DNA sequence encoding results in what is called sparse data. That is, the majority of the _features_ are empty. Scaling and transforming sparse and or ordinal data will __destroy__ the sparse structure within the data, and is not recommended.


### Scaling and transforming the target vector

With many real world regression problems, you will find that the target vector is imbalanced and or skewed.

For example, the DIS feature in the Boston dataset is imblanaced or skewed. If we were to use this as a regression target, we would find the model to be less reliable towards the larger values.

````python
dis_data = boston.data[:,7] # Extract the DIS feature data

plt.figure(figsize=(6, 6))
plt.hist(dis_data, bins=50)
plt.xlabel('DIS')
plt.ylabel('Count')
plt.show()
````

Further, some machine learning models struggle with regression values that are significantly greater than 1.0

As such, it is common to transform and scale the target vector in regression problems. This does not apply to classification problems.

Log transformation of the target vector prior to scaling may result in improved model accuracies, as the resulting data resembles more of a normal distribution.

````python
dis_data_log = np.log(dis_data)

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.hist(dis_data, bins=50)
plt.xlabel('DIS')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(dis_data_log, bins=50)
plt.xlabel('DIS')
plt.ylabel('Count')
plt.show()
````

This is particularly problematic when dealing with rare events such as financial/credit card fraud.

### Missing values

In many cases, you may come across a dataset with missing feature/target values.

Here, there are a few approaches you can take:

* Remove all rows/samples with missing values.
* Fill in missing values with the mean or median of that feature.
* Fill in missing values with 0 or 1, if appropriate.
* Train a model on data without missing data, and use it to predict the missing values.
* Collect more/better data.

### Principal Component Analysis (PCA)

If you recall from the Boston housing dataset with 13 features, the plots were pretty unweildy. By applying PCA to the dataset, we can extract out the principal components, or the components that are most correlated with the data.

> Keep in mind that PCA cannot be used on sparse or ordinal data. There must be meaning in the distances between values. For sparse/ordinal data such as nucleic and protein sequences, use `TruncatedSVD` or some of the more advanced methods.

> Further, PCA requires your data to be scaled.

As a side note, it is also possible to use PCA transformed data as an input to most machine learning algorithms.

````python
# PCA on raw boston data

from sklearn.decomposition import PCA

boston_pca = PCA(n_components=2)
boston_pca_tfm = boston_pca.fit_transform(boston.data)

plt.figure(figsize=(10,7))
plt.scatter(boston_pca_tfm[:, 0], boston_pca_tfm[:, 1], c=boston.target, cmap=sns.cubehelix_palette(2, start=0.5, rot=-0.5, as_cmap=True, reverse=True))

plt.colorbar()
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")

plt.show()
````

### t-distributed Stochastic Neighbor Embedding (t-SNE)

t-SNE is a tool to visualize high-dimensional data. It converts similarities between data points to joint probabilities and tries to minimize the Kullback-Leibler divergence between the joint probabilities of the low-dimensional embedding and the high-dimensional data.

It is highly recommended to use another dimensionality reduction method (e.g. PCA for dense data or TruncatedSVD for sparse data) to reduce the number of dimensions to a reasonable amount (e.g. 50) if the number of features is very high.

````python
from sklearn import manifold
from sklearn import preprocessing

standard_scaler = preprocessing.StandardScaler()
boston_data_scaled = standard_scaler.fit_transform(boston.data)

boston_tsne = manifold.TSNE(n_components=2, perplexity=100, init='random')
boston_tsne_tfm = boston_tsne.fit_transform(boston_data_scaled)

plt.figure(figsize=(10,7))
plt.scatter(boston_tsne_tfm[:, 0], boston_tsne_tfm[:, 1], c=boston.target, cmap=sns.cubehelix_palette(2, start=0.5, rot=-0.5, as_cmap=True, reverse=True))

plt.colorbar()
plt.title('t-SNE of standardised Boston data.')

plt.show()
````

### K-means clustering

One of the most common and simplest clustering methods is [K-means clustering](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans).

Briefly, the K-means algorithm clusters data by trying to separate samples in n groups of equal variance, minimizing a criterion known as the inertia. This algorithm requires the number of clusters to be specified. It scales well to large number of samples and has been used across a large range of application areas in many different fields.

The example below will show you have effective it is at grouping a 'known' number of clusters from a two-dimensional dataset.

````python
from sklearn import cluster
from numpy.random import normal, shuffle

def cluster_scatter(plot, data, labels):
    x_vals, y_vals = np.hsplit(data, 2)
    
    for k in range(labels.max()+1):
        idx = labels == k
        plot.scatter(x_vals[idx], y_vals[idx], alpha=0.5)  


# Some random data in three blobs
std  = 0.25
n = 100
data = [normal((0.0, 0.0), std, (n,2)),
        normal((1.0, 1.0), std, (n,2)),
        normal((1.0, 0.0), std, (n,2))]

data = np.vstack(data)
orig_labels = np.arange(0,3,1/n).astype(int)

plt.figure(figsize=(7,5))
x_vals, y_vals = np.hsplit(data, 2)
plt.scatter(x_vals, y_vals)
plt.show()
````

````python
# Show the class labels of the blobs.

plt.figure(figsize=(7,5))
cluster_scatter(plt, data, orig_labels)
plt.show()
````

````python
# Train a Kmeans model on the unlabelled data.

model = cluster.KMeans(3)
labels = model.fit_predict(data)
````

````python
# Compare the True cluster labels, with those predicted by KMeans.
fig, (sp1, sp2, sp3) = plt.subplots(1,3)

cluster_scatter(sp1, data, orig_labels)
sp2.scatter(x_vals, y_vals)
cluster_scatter(sp3, data, labels)

sp1.set_title('True clusters')
sp2.set_title('Input to K-means')
sp3.set_title('K-means')

plt.show()
````

````python
# Need to know or have a good estimate of the number of clusters.

model = cluster.KMeans(4)
labels = model.fit_predict(data)

plt.figure(figsize=(7,5))
cluster_scatter(plt, data, labels)
plt.show()
````

## Supervised learning - Regression


As mentioned in the introduction, regression problems are those in which we wish to predict a continuous value for a given set of inputs. Many of the Scikit-learn regression models can only deal with one-dimensional data, that is, a list of features.

Images, which are > 2 dimensions need to be reduced or flattened to a list of numbers.


Scikit-learn implements a very large number of [supervised learning algorithms](https://scikit-learn.org/stable/supervised_learning.html#supervised-learning), many of which have both regression and classification versions.

__Evaluating__ regression models can be done via a few methods, many of which are implemented by [Scikit-learn](https://scikit-learn.org/stable/modules/model_evaluation.html).

__[MSE: Mean Squared Error](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html#sklearn.metrics.mean_squared_error)__, measures average squared error of the predictions. For each sample, it calculates square difference between the predictions and the target and then average those values. The higher this value, the worse the model is. It is never negative.

However, if we make a single very bad prediction, the squaring will make the error even worse and it may skew the metric towards indicating that the model is worse than it is. In such a case, use MAE.

__[MAE: Mean Absolute Error](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_absolute_error.html#sklearn.metrics.mean_absolute_error)__ is calculated as an average of absolute differences between the target values and the predictions. MAE is a linear score which means that all the individual differences are weighted equally in the average. This means that large once off errors do not affect the MAE as much as MSE.

__[R^2 or coefficient of determination](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html#sklearn.metrics.r2_score)__ measures the 'goodness of fit' of the model to the data. 0 is bad, 1 is perfect fit to the data. It is similar to MSE, but is scale free, always giving a result between -inf and +1.0.

__Plotting true vs. predicted values__ can give you a visual assessment of the model's predictive power over the range of the target.


### Using Scikit-learn models

````python
# Prepare the Boston housing dataset.

from sklearn.model_selection import train_test_split
from sklearn import preprocessing

X = boston.data
y = boston.target

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the training data and then use the same parameters to scale the test set.
standard_scaler = preprocessing.StandardScaler()

X_train = standard_scaler.fit_transform(X_train)
X_test = standard_scaler.transform(X_test)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
````

### [Linear regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html)

The Scikit-learn implementation of `LinearRegression` fits a linear model with coefficients for each feature, using Ordinary Least Squares.

This model is fast to train and always serves as a good baseline.

````python
from sklearn.linear_model import LinearRegression

# Create the model.
lr = LinearRegression()

# Fit the model with our training dataset.
lr.fit(X_train, y_train)

# Use the model to predict our test set.
predicted = lr.predict(X_test)
expected = y_test
````

````python
# Evaluate the models predictions on the test set with a series of metrics.

from sklearn.metrics import mean_squared_error, mean_absolute_error

print("Train set R^2: ", lr.score(X_train, y_train))
print("Test set R^2: ", lr.score(X_test, y_test))
print("Test set MSE: ", mean_squared_error(expected, predicted))
print("Test set MAE: ", mean_absolute_error(expected, predicted))
````

````python
# Plot the predicted target values against the true target values.

plt.figure(figsize=(7,5))
plt.scatter(expected, predicted)
plt.plot([0, 50], [0, 50], '--k')
plt.xlabel('True price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
plt.show()
````

### [Nearest Neighbors Regression](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html)

The Scikit-learn implementation of `KNeighborsRegressor` uses â€˜feature similarityâ€™ to predict values of any new data points.

This means that the a sample is assigned a value based on how closely it resembles the points in the training set.

More detail can be found in the Scikit-learn documentation.

````python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors=2).fit(X_train, y_train)

predicted = knr.predict(X_test)
expected = y_test

print("Train set R^2: ", knr.score(X_train, y_train))
print("Test set R^2: ", knr.score(X_test, y_test))
print("Test set MSE: ", mean_squared_error(expected, predicted))
print("Test set MAE: ", mean_absolute_error(expected, predicted))

plt.figure(figsize=(7,5))
plt.scatter(expected, predicted)
plt.plot([0, 50], [0, 50], '--k')
plt.xlabel('True price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
plt.show()
````

### [Support Vector Machines (SVM)](https://scikit-learn.org/stable/modules/svm.html)

Support Vector Machines are typically a linear model that is capable of learning a non-linear function by use of something called the [kernel trick](https://en.wikipedia.org/wiki/Kernel_method), which projects or transforms the data into another dimension where there is a clear linear separation/fit to the data.


````python
from sklearn import svm

# svr = svm.SVR().fit(X_train, y_train)
svr = svm.SVR(gamma='auto', C=10.0, epsilon=0.1).fit(X_train, y_train)

predicted = svr.predict(X_test)
expected = y_test

print("Train set R^2: ", svr.score(X_train, y_train))
print("Test set R^2: ", svr.score(X_test, y_test))
print("Test set MSE: ", mean_squared_error(expected, predicted))
print("Test set MAE: ", mean_absolute_error(expected, predicted))

plt.figure(figsize=(7,5))
plt.scatter(expected, predicted)
plt.plot([0, 50], [0, 50], '--k')
plt.xlabel('True price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
plt.show()
````

### [Random forests](https://scikit-learn.org/stable/modules/ensemble.html#random-forests)

A random forest is an ensemble estimator that fits a number of [decision trees](https://en.wikipedia.org/wiki/Decision_tree) on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.

Another advantage of tree based methods, is that they are very interpretable. They can tell you exactly why they made a decision.

````python
from sklearn.ensemble import RandomForestRegressor

regr = RandomForestRegressor(max_depth=5, random_state=0, n_estimators=100).fit(X_train, y_train)

predicted = regr.predict(X_test)
expected = y_test

print("Train set R^2: ", regr.score(X_train, y_train))
print("Test set R^2: ", regr.score(X_test, y_test))
print("Test set MSE: ", mean_squared_error(expected, predicted))
print("Test set MAE: ", mean_absolute_error(expected, predicted))

plt.figure(figsize=(7,5))
plt.scatter(expected, predicted)
plt.plot([0, 50], [0, 50], '--k')
plt.xlabel('True price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
plt.show()
````

### [Gradient Boosted Trees](https://scikit-learn.org/stable/modules/ensemble.html#gradient-boosting)

Gradient boosted trees are also an ensemble method, and are similar to random forests, in that they build numerous decision trees and average their predictions.

However, they differ from random forests in that they build their trees iteratively, adjusting parameters based on the performance of the previous decision tree.

Gradient boosted classifiers and regressors tend to be some of the most powerful classical machine learning (non-neural network) methods available.

````python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)

predicted = gbr.predict(X_test)
expected = y_test

print("Train set R^2: ", gbr.score(X_train, y_train))
print("Test set R^2: ", gbr.score(X_test, y_test))
print("Test set MSE: ", mean_squared_error(expected, predicted))
print("Test set MAE: ", mean_absolute_error(expected, predicted))

plt.figure(figsize=(7,5))
plt.scatter(expected, predicted)
plt.plot([0, 50], [0, 50], '--k')
plt.xlabel('True price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
plt.show()
````

### Feature importances

### [ROC curve](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_curve.html#sklearn.metrics.roc_curve)

A useful tool when predicting the probability of a binary outcome is the Receiver Operating Characteristic curve, or ROC curve.

It is a plot of the false positive rate (x-axis) versus the true positive rate (y-axis) for a number of different candidate threshold values between 0.0 and 1.0.

More specifically, __it plots the false positive rate versus the hit rate.__

The ROC curve is a useful tool for a few reasons:

* The curves of different models can be compared directly in general or for different thresholds.
* The area under the curve (AUC) can be used as a summary of how good the model is.


The shape of the curve contains a lot of information:

* Smaller values on the x-axis of the plot indicate lower false positives and higher true negatives.
* Larger values on the y-axis of the plot indicate higher true positives and lower false negatives.

___However, ROC curves should only be used when there are roughly equal numbers of observations for each class.___

__If the classes are imbalanced, use a [Precision-Recall curve](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_recall_curve.html#sklearn.metrics.precision_recall_curve) instead.__

````python
# Generate a ROC AUC plot
import sklearn.metrics as metrics

fpr, tpr, threshold = metrics.roc_curve(y_test, predicted)
roc_auc = metrics.auc(fpr, tpr)

plt.figure(figsize=(7,5))
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, label = 'AUC = %0.2f' % roc_auc)
plt.plot([0, 1], [0, 1],'r--', label = 'Random chance')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc = 'lower right')
plt.show()
````

### [Precision-Recall curve](https://scikit-learn.org/stable/auto_examples/model_selection/plot_precision_recall.html#sphx-glr-auto-examples-model-selection-plot-precision-recall-py)

The precision-recall curve shows the tradeoff between precision and recall for different thresholds.

A high area under the curve represents both high recall and high precision, where high precision relates to a low false positive rate, and high recall relates to a low false negative rate.

High scores for both show that the classifier is returning accurate results (high precision), as well as returning a majority of all positive results (high recall).

````python
# Generate a PR plot
import sklearn.metrics as metrics

precision, recall, _ = metrics.precision_recall_curve(y_test, predicted)

plt.figure(figsize=(7,5))
plt.step(recall, precision, where='post')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.show()
````

### [Confusion matrix](https://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html)

A confusion matrix is a table that is often used to describe the performance of a classification model on a set of test data for which the true values are known.

|    .      |Pred: Yes|Pred: No|
| ---     | ---  | ---  |
| __True: Yes__|TP       | FN     |
| __True: No__ |FP       | TN     |


__* True positives (TP):__ We predicted yes for disease, and they do have the disease.

__* True negatives (TN):__ We predicted no, and they don't have the disease.

__* False positives (FP):__ We predicted yes, but they don't actually have the disease. (Also known as a "Type I error.")

__* False negatives (FN):__ We predicted no, but they actually do have the disease. (Also known as a "Type II error.")



Whilst the confusion matrix is a great visualisation of what the classifier is doing, if the classes are imbalanced, it will give you a false sense of accuracy.

````python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(expected, predicted)

plt.figure(figsize=(7,5))
ax = plt.subplot()
sns.heatmap(cm, annot=True, ax = ax, cmap=plt.cm.Blues, fmt='g')

# labels, title and ticks
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['Malignant', 'Benign'])
ax.yaxis.set_ticklabels(['Malignant', 'Benign'])

plt.show()
````

## Training and hyperparameter optimisation

As I briefly mentioned, and as you might have seen, many of the machine learning algorithms implemented by Scikit-learn have default and optional parameters that are set prior to learning.

These parameters, called [hyperparameters](https://en.wikipedia.org/wiki/Hyperparameter_(machine_learning)), control various aspects of how the algorithms fit to the data.

Some simple algorithms, such as [ordinary least squares regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression), have no hyperparameters, whilst [LASSO](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html#sklearn.linear_model.Lasso) is basically OLS with an extra [regularisation](https://en.wikipedia.org/wiki/Regularization_(mathematics)) term, that helps mitigate overfitting.

More advanced machine learning algorithms will have more hyperparameters that control things like the learning rate, number of learning cycles, regularisation weights, number of estimators to create, depth of tree, etc.

Adjusting these parameters for identifying the best performing model is a process known as [hyperparameter optimisation](https://en.wikipedia.org/wiki/Hyperparameter_optimization).

Scikit-learn implements a [few methods for hyperparameter optimisation](https://scikit-learn.org/stable/modules/grid_search.html), such as grid search and random search. Further, certain models come with their own model specific version that can automatically [tune hyperparameters](https://scikit-learn.org/stable/modules/grid_search.html#alternatives-to-brute-force-parameter-search).

## Ensemble methods

In a typical machine learning pipeline, one will generally evaluate a large number of models on the dataset, followed by hyperparameter optimisation on all or several of the best performing models.

However, at some point, you will hit an upper limit to the performance of that particular algorithm on the given training dataset.

__A very simple and effective trick__ at extracting more performance is by combining several of the best models in something called an ensemble. That is, you take the average/median/consensus prediction of several models.

Scikit-learn implements these as [Voting Classifiers](https://scikit-learn.org/stable/modules/ensemble.html#voting-classifier) and [Voting Regressors](https://scikit-learn.org/stable/modules/ensemble.html#voting-regressor).

By combining several of the same or different models will typically result in a more accurate (3-5%) and robust (reduce overfitting) predictor with essentially no extra work.
